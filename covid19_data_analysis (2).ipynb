```python
#import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
#delete the useless column
df = dataset.drop(["Lat","Long"], axis = 1, inplace = True)
```

```python
#aggregate the rows by the country
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
# Assuming corona_dataset_aggregated is your DataFrame
# Convert any non-numeric columns to numeric, excluding the index
corona_dataset_aggregated = corona_dataset_aggregated.apply(pd.to_numeric, errors='coerce')

# Extract list of countries from the DataFrame index
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries:
    # Calculate the maximum infection rate for each country
    max_rate = corona_dataset_aggregated.loc[c].diff().max()
    max_infection_rates.append(max_rate)

# Add the max_infection_rates as a new column to the DataFrame
corona_dataset_aggregated["Max_infection_rates"] = max_infection_rates
```

```python
#create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
```
